## Vendas das lojas resumido

Agrupado por categoria

```python
import json
import pandas as pd

categories = json.load(open('subcategorias.json'))
regions = json.load(open('regioes.json'))

stores = []
for region in regions:
    stores += regions[region]
    
index = pd.date_range(start='2018-01-01', end='2021-04-01', freq='D')
columns = ['Teste'] # Coluna provisória
```

```python
df_out = pd.DataFrame(index=index, columns=columns)
df_out.fillna(value=0, inplace=True)

i = 0
for store in stores:
    i += 1
    print('(%d/%d): %s' % (i, len(stores), store))
    # Abrir o .csv das vendas da loja
    df = pd.read_csv('data/' + store + '.csv', index_col=0)
    df.index = pd.to_datetime(df.index)
    # Obter as vendas por categoria da loja
    store_columns = []
    for category in categories:
        store_columns.append(store + "|" + category)
        df_out[store + "|" + category] = df[categories[category]].sum(axis=1)
    # Obter a venda de tudo da loja
    df_out[store + "|GERAL"] = df_out[store_columns].sum(axis=1)
```

```python
for region in regions:
    columns = []
    for store in regions[region]:
        for category in categories:
            columns.append(store + '|' + category)
    df_out[region + "|GERAL"] = df_out[columns].sum(axis=1)
```

```python
for region in regions:
    for category in categories:
        columns = []
        for store in regions[region]:
            columns.append(store + '|' + category)
        df_out[region + "|" + category] = df_out[columns].sum(axis=1)    
```

```python
df_out.to_csv('Resumo Lojas.csv')
```
